```python
# 간단한 sin 그래프
x = np.linspace(0, 9.5, 1000)
fig = plt.figure()
ax = plt.axes()
sin = ax.plot(x, np.sin(x), 'r')
cos = ax.plot(x, np.cos(x), 'b')
tan = ax.plot(x, np.tan(x/2), 'g')
ax.set_xlim(0,9.5)
ax.set_ylim(-5,5)
ax.grid()
```

```python
x = np.linspace(0, 9.5, 1000)
fig = plt.figure()
ax = plt.axes()
sin = ax.plot(x, np.sin(x), 'r', label='사인')
cos = ax.plot(x, np.cos(x), 'b', label='코사인')
tan = ax.plot(x, np.tan(x/2), 'g', label='탄젠트')
ax.set(xlim=(0,9.5), ylim=(-1.1,1.1), title='sin/cos/tan 그래프', xlabel='x값',ylabel='y값')
ax.legend()
ax.grid()
```

### 여러개의 그래프 나타내기 : subplot
+ subplot(행수/열수/번호)

```python
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)

y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

# 2x1 그래프 
fig = plt.figure()
ax1 = plt.subplot(211)
plt.plot(x1,y1, 'ro--')

ax2 = plt.subplot(212)
plt.plot(x2,y2, 'b*:')
```

```python
# 1 x 2 그래프
fig = plt.figure()
ax1 = plt.subplot(121)
plt.plot(x1,y1, 'ro--')

ax2 = plt.subplot(122)
plt.plot(x2,y2, 'b*:')
```

```python
# 2 x 2 그래프
plt.subplot(221) # 1,1
plt.plot(np.random.rand(10), 'r')

plt.subplot(222) # 1,2
plt.plot(np.random.rand(10), 'b')

plt.subplot(223) # 2,1
plt.plot(np.random.rand(10), 'g')

plt.subplot(224) # 2,2
plt.plot(np.random.rand(10), 'k')
```

```python
# 2 x 2 그래프2
fig, axes = plt.subplots(2,2)
axes[0,0].plot(np.random.rand(10), 'r')
axes[0,1].plot(np.random.rand(10), 'b')
axes[1,0].plot(np.random.rand(10), 'g')
axes[1,1].plot(np.random.rand(10), 'k')

axes[0,0].set_title('axes1')
axes[0,1].set_title('axes2')
axes[1,0].set_title('axes3')
axes[1,1].set_title('axes4')

plt.tight_layout() # 여러개의 그래프를 그리는 경우
```

```python
# 부산 -> 서울, 경기, 인천 인구 이동을 시각화 해보세요
popmove = pd.read_excel('csv/시도별_전출입_인구수.xlsx', header=0)
popmove = popmove.fillna(method='ffill')
popmove.head()
```

```python
# 데이터 추출
serach = (popmove.전출지별 == '부산광역시') & (popmove.전입지별 == '서울특별시')
psmove = popmove[serach]

serach = (popmove.전출지별 == '부산광역시') & (popmove.전입지별 == '경기도')
pkmove = popmove[serach]

serach = (popmove.전출지별 == '부산광역시') & (popmove.전입지별 == '인천광역시')
pimove = popmove[serach]
#pimove = pimove.replace('-',np.NaN)
```

```python
x = np.arange(2000,2017+1)
psy = psmove.iloc[:,32:]
psy = psy.iloc[0,]

pky = pkmove.iloc[:,32:]
pky = pky.iloc[0,]

piy = pimove.iloc[:,32:]
piy = piy.iloc[0,]
```

```python
plt.plot(x,psy, 'r', label='부산-서울')
plt.plot(x,pky, 'b', label='부산-경기')
plt.plot(x,piy, 'g', label='부산-인천')
plt.legend()
```

```python
# 그래프로 시각화2
fig = plt.figure()
ax = plt.axes()
plt.plot(x,psy, 'r', label='부산-서울')
plt.plot(x,pky, 'b', label='부산-경기')
plt.plot(x,piy, 'g', label='부산-인천')
ax.legend()

# x/y축 눈금을 리스트로 변환 후 포맷문자열로 변환
ax.set_xticklabels(['{:}'.format(int(x)) for x in ax.get_xticks().tolist()])
ax.set_yticklabels(['{:,}'.format(int(y)) for y in ax.get_yticks().tolist()])
plt.show()
```

```python
# 그래프로 시각화3
plt.subplot(2,1,1)
plt.plot(x,psy,'r')
plt.title('부산->서울')
plt.xlim(2000,2017)

ax2 = plt.subplot(2,2,3)
plt.plot(x,pky,'b')
plt.title('부산->경기')
plt.xlim(2000,2017)

ax3 = plt.subplot(2,2,4, sharey=ax2)
plt.plot(x,piy,'m')
plt.title('부산->인천')
plt.xlim(2000,2017)
plt.yticks(visible=False)

plt.tight_layout()
plt.show()
```
