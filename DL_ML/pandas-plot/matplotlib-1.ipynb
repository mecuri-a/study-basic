```python
np.random.seed(2111021105)
y = np.random.standard_normal(20) # 정규분포 난수 20개
x = np.arange(1,20+1)
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
# 암시적 스타일 지정 : 색 // r ,g ,b, c, m, y, k, w
plt.plot(x,y,'r')
```

```python
# 암시적 스타일 지정 : 마커(데이터를 나타내는 기호 //  . / o / v / ^ / 1 /p /* / D
plt.plot(x,y,'cD')
```

```python
# 암시적 스타일 지정 : 선모양 //  실선(-), 대시선(:), 점선(--), 대시점선(-.)
plt.plot(x,y,'c--')
```

```python
# 명시적 스타일 지정 : color(c), linestyle(ls), linewidth(lw), marker, markersize(ms), markerfacecolor(mfc), markeredgecolor(mec)
plt.plot(x,y, c='m', ls='-.', lw=3, marker='d', ms=10, mfc='k', mec='k')
```

```python
# 축 지정하기 : xlim(최소, 최대), ylim(최소, 최대)으로 설정
plt.plot(x,y)
plt.xlim(-1,21)
plt.ylim(-3,3)
```

### 축 눈금 지정하기
+ 축 상의 눈금 tick 이라 함
+ xticks(수치리스트, 레이블)
+ yticks(수치리스트, 레이블)
+ linspace : 지정한 범위를 특정 수 만큼 일정하게 나눔
    - np.linspace(1,10,10)
    - np.linspace(0,1,10)

```python
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xticks([-3.14,0,3.14])
plt.yticks([-1,-0.5,0,0.5,1])
plt.show()
```

```python
# 눈금대신 레이블로 대신 출력할 수도 있음
plt.plot(x,y)
plt.xticks([-3.14,-3.14/2,0,3.14/2,3.14],['a','b','c','d','e'])
plt.yticks([-1,-0.5,0,0.5,1],['가','나','다','라','마'])
plt.show()
```

```python
# 안내선 출력 : grid line
# 범례 표시
#자동으로 범례를 표시하려면 각 그래프 작성시 label로 이름을 지정해야 함
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1, 'r--',label='코사인')
plt.plot(x,y2, 'b--',label='사인')
plt.legend(loc='best')
```
