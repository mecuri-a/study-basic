### 데이터 시각화
* 데이터를 관찰하는 과정에서 선택사항이 아니라 반드시 거쳐야 하는 필수 과정
* 숫자와 텍스트만으로 구성된 통계결과는 지루/따분
* 그래프와 그림형태로 제시하면 데이터의 관계구분 용이, 열정적인 반응 동반

### R 기본 그래프 함수
* plot, hist, boxplot, barplot, pie, stem

```r
# 표준정규분포를 따르는 난수 15개 생성
# 그에 따른 그래프를 작성해서 시각화
# 정규분포 : 수집한 자료가 평균 0, 표준편차가 1인 형태
# rnorm(갯수) : -1 ~ 1 사이의 난수 생성 함수
x <- rnorm(15, mean=0, sd=1)  # 안됨 ㅋㅋ
summary(x)
```

```r
# 산점도 : plot(x,y,옵션)
plot(x)        # 변수가 1개 일때는 자동으로 y축에 할당
```

```r
# 막대그래프 : barplot(x,y,옵션)
barplot(x)
```

```r
# 히스토그램 hist(x, 구간, 옵션)
hist(x)
```

```r
# 상자수염그래프 : boxplot(x,옵션)
boxplot(x)
```

```r
# 줄기잎그래프 : stem(x,옵션)
stem(x)
```

```r
# 두 약물에 따른 환자반응 데이터
dose <- c(20,30,40,45,60)   # 나이
drugA <- c(16,20,27,40,60) # 효과 정도
drugB <- c(15,18,25,31,40)

# 학생 10명의 키 데이터
hights <- c(170,165,182,156,142,165,159,147,175,172)
```

## 산점도
+ n개의 짝으로 이루어진 자료(컬럼이 2개이상)를 x,y 평면에 점으로 나타내는 그래프
+ 자료의 본포정도를 파악하는데 사용
+ 주로 상관, 회귀 분석에 사용
+ plot(x축값, y축값, 그래프옵션)
+ plot(y축값, 그래프옵션) => x축 자동지정

```r
# 단순 출력
plot(hights)
```

```r
plot(x=c(11:20), y=hights)
```

```r
plot(dose, drugA)
```

```r
plot(dose, drugB)
```

### 그래프 다양한 설정 보기
+ main : 제목
+ sub : 부제목
+ xlab/ylab : 축제목
+ axes : x/y축 제목
+ xlim/ylim : 축값 범위
+ col : 그래프 색상
+ type : 그래프 유형  l -> 라인, p -> 점, b -> 둘다

```r
plot(hights, main='student hights', sub='class A', xlab='numbers', ylim=c(140,190), col='red', type='b')
```

```r
# 산점도 점문자 유형(pch) 
plot(hights, main='student hights', sub='class A', xlab='numbers', ylim=c(140,190), col='black', type='b', pch=10)  # pch=숫자  -> 형태 변환 1 기본, 2 세모....
grid() # 모눈 표시
```

### 그래프 눈금 새부 조정하기
+ 그래프 그릴때 axes 속성을 FALSE로 지정
+ axis 함수로 세부 조정
+ axis(축지정, 눈금지정)

```r
plot(hights, main='student hights', sub='class A', xlab='numbers', col='blue', pch=3, axes=F)
axis(side=1, at=1:10)           # side =1 x축 설정
axis(side=2, at=seq(140,180,5), las=2) # side = 2, y축 설정  , las = y축 값을 수평 방향으로 바꿔줌
box()                           # 그래프 테투리 표시
grid()
```

### 그래프 텍스트 추가하기
+ 특정 x,y위치에 글자 표시
+ text(y좌표, 글자크기(cex), 글자위치(adj), 글자내용(lab))
+ adj : 음수 -> (오른쪽,위), 양수 -> (왼쪽, 아래)

```r
plot(hights, col='red')
text(hights, cex=0.7, adj=c(0.5, -1), lab=hights)
```

```r
# A 약물에 따른 환자반응 - 눈금조정, 텍스트추가
plot(dose, drugA, axes=F)
axis(side=1, at=1:70)
axis(side=2, at=1:70, las=2)
box()|grid()
text(drugA, cex=1, adj=c(0.5,-1), lab=drugA)
```

```r
# iris 데이터셋에서 Petal Length/Width로 시각화
str(iris)
plot(iris$Petal.Length, iris$Petal.Width)
```

```r
# 독립변수에 따른 종속변수도 시각화 붓꽃의 품종을 o, △, + 기호로 표현
plot(iris$Petal.Length, iris$Petal.Width, pch=as.integer(iris$Species))
```

```r
# 독립변수에 따른 종속변수도 시각화 붓꽃의 품종을 색상으로 표현
# 1:black, 2:red, 3:green, 4: blue
plot(iris$Petal.Length, iris$Petal.Width, pch=as.integer(iris$Species),col=as.integer(iris$Species)+1)
```

### 범례 표시하기
+ legend(위치, 값, 기호, 색상)
+ 위치 : top, bottom, right, left, center

```r
plot(iris$Petal.Length, iris$Petal.Width, pch=as.integer(iris$Species),col=as.integer(iris$Species)+1)
legend('topleft', c('setosa','versicolor','verginica'), pch=c('o','△','+'), col=c('red','green','blue'))
```

```r
# 자동으로 텍스트, 기호, 색상 지정
plot(iris$Petal.Length, iris$Petal.Width, pch=as.integer(iris$Species),col=as.integer(iris$Species)+1)
legend('bottomright', levels(iris$Species), pch=unique(as.integer(iris$Species)), col=unique(as.integer(iris$Species))+1)
```

### 2개의 산점도를 동시에 출력
+ plot함수와 lines 함수를 번갈아 가며 사용
+ 이런 경우 범례를 반드시 출력할 것!

```r
# A,B 약물에 따른 환자반응
plot(dose, drugA, col='red', type='b', pch=1, ylab='drugs')
lines(dose, drugB, col='blue', type='b', pch=2)
legend('topleft', c('drugA','drugB'), pch=c(1,2), col=c(2,1))
```
