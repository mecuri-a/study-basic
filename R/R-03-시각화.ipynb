### 막대그래프
+ 비율 척도상의 크기를 막대형태로 나타낸 것
    - 한편, 등간척도를 이용한 돗수분포표를 막대그래프로 나타내기도 함
+ 주로 자료의 범주별 빈도를 요약해서 나타냄
+ barplot(대상, 누적여부, 표시방향)

```r
load('rdata/height.rdata')
str(height)
```

```r
# 막대그래프 시각화
# names.arg : x축 레이블 표시, 대상은 벡터로 정의
barplot(height, names.arg=c(1:80), col=rainbow(10))
```

```r
# horiz = T 가로
barplot(height, names.arg=c(1:80), col=rainbow(10), horiz=T)
```

```r
# text 함수로 막대그래프에 숫자 표시
bp <- barplot(height, names.arg=c(1:80), col=rainbow(10), axes=F)
axis(side=2, at=seq(0,200,10))
text(bp, height, lab=height)
```
